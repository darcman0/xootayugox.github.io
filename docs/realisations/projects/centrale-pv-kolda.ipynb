# Levé topo — Centrale photovoltaïque Kolda

## Contexte

Mission réalisée avec Casa Topo en août-septembre 2024 à Kolda. Étude topographique par levé drone pour l'installation d'une centrale photovoltaïque.

## Données clés

| Indicateur | Valeur |
|---|---|
| Surface levée | 330 ha |
| Centrale PV | 70 ha |
| Ligne d'évacuation | 10,5 km |
| Drone | Phantom 4 Pro |
| Logiciel | Pix4D + Metashape |

## Méthode

- Mise en place des GCP avec GPS différentiel
- Planification des missions avec DroneDeploy
- Extraction MNT, MNS et orthophotos
- Identification des éléments d'occupation du sol

## Résultats

Livrables MNT, MNS et orthophotos haute résolution transmis au client pour la conception de la centrale.
