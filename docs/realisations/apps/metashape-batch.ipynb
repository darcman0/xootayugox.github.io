# Metashape Batch Processor

Script d'automatisation du traitement photogrammétrique dans Agisoft Metashape Professional.

## Contexte

Développé lors de la mission WeForest — 300 sites de mangrove à traiter. Ce script réduit de 20% le temps de traitement en éliminant les étapes manuelles répétitives.

## Fonctionnalités

- Alignement des photos en haute précision
- Génération du nuage de points dense
- Export orthophoto à 2 cm de résolution
- Traitement batch sur plusieurs dossiers

```python
import Metashape
import os

doc = Metashape.app.document
chunk = doc.chunk

# Alignement
chunk.matchPhotos(accuracy=Metashape.HighAccuracy)
chunk.alignCameras()

# Nuage dense
chunk.buildDepthMaps(quality=Metashape.MediumQuality)
chunk.buildDenseCloud()

# Orthophoto 2 cm
chunk.buildOrthomosaic(resolution=0.02)
chunk.exportOrthomosaic('output_ortho.tif', resolution=0.02)

print('Traitement terminé :', chunk.label)
```
